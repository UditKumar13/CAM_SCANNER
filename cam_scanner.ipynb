```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
% matplotlib inline
```

```python
im_path='E:/Github_projects/Cam_Scanner/receipt.jpg'
#boundary: contour 
im=cv2.imread(im_path)
print(im.shape)
img=cv2.resize(im,(1500,1000))
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
## img blurring 
# to detect boundares we have to blurr it first 
#blurr works on a gray scale : store intensity of each pixels 
orig=img.copy()
gray=cv2.cvtColor(orig,cv2.COLOR_BGR2GRAY)
blurred=cv2.GaussianBlur(gray,(5,5),0)
# here 0 is  a sigma standard deviation 
plt.imshow(gray)
plt.show()
print("---------this is gray scale image-------- ")
plt.imshow(blurred)
plt.show()
print("---------this is blurred  scale image-------- ")
print(gray.shape)
print("---gray shape----")
print(blurred.shape)
print("---blurred shape----")
```

```python
#edge detection   after image smoothing 
#we have a direct function canny which also run a kernel for edge detection 
edged=cv2.Canny(blurred ,0,50)
#0 and 50 are thresold 1,thresold 2
#check the query 
# cv2 .canny 
orig_edged=edged.copy()

plt.imshow(orig_edged)
plt.show()
```

```python
#contour extraction 
#contours are the ploygon on the image ,which  return multiple values
# intensity value is linear equation with rgb
#-IMAGE,CONTOURS,-HIERARCHY
_,contours,_=cv2.findContours(edged,cv2.RETR_LIST,cv2.CHAIN_APPROX_NONE)
print(len(contours))
contours=sorted(contours,reverse=True,key=cv2.contourArea)
```

```python
print(contours[2])
```

```python
## now we will select best contour from all contours 
for c in contours:
    p=cv2.arcLength(c,True)
    approx=cv2.approxPolyDP(c,p*0.02,True)
    # 2% arclength works well 
    
    if len(approx) == 4:
        target=approx
        break
print(target.shape)
print(target)
```

```python
t=[[1,2],[2,2],[3,3],[43,1]]

t=np.array(t)
print(t)
t=t.sum(axis=1)
## adding each row elements 
print(t)
```

```python
## reorder the target contour function 
def reorder(h):
    h=h.reshape((4,2))
    print(h)
    
    
    ## target is our input contour 
    hnew=np.zeros((4,2),dtype=np.float32)
    add =h .sum(axis=1)
    hnew[3]=h[np.argmin(add)]
    hnew[1]=h[np.argmax(add)]
    
    diff=np.diff(h,axis=1)
    hnew[0]=h[np.argmin(diff)]
    hnew[2]=h[np.argmax(diff)]
    
    return hnew
```

```python
reordered=reorder(target)
print("*****----reordered target contour----********")
print(reordered)
```

```python
## let us now project to a fixed size-screen

## let us make a transformtion matrix 
input_represent=reordered
output_map=np.float32([[0,0],[1000,0],[1000,1000],[0,1000]])
## 900 is taken due to photo size
M=cv2.getPerspectiveTransform(input_represent,output_map)
#M is a 3*3 matrix
print(M)
```

```python
ans=cv2.warpPerspective(orig,M,(800,800))
## its first argument remain our actual image and second argument is perspective transform matrix
```

```python
plt.imshow(ans)
plt.show()
```

```python
ans2 = cv2.cvtColor(ans, cv2.COLOR_BGR2GRAY)
temp = cv2.GaussianBlur(ans2, (3,3), 0)
temp2 = cv2.GaussianBlur(ans2, (7,7), 0)
temp3 = cv2.GaussianBlur(ans2, (1,1), 0)
## more poolishing of image 
plt.imshow(temp)
plt.show()

plt.imshow(temp2)
plt.show()

plt.imshow(temp3)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
% matplotlib inline
```

```python
im_path='E:/ML/screen_pics/9. CAM SCANNER/scan_form.jpg'
#boundary: contour 
im=cv2.imread(im_path)
print(im.shape)
img=cv2.resize(im,(900,1300))
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
## img blurring 
# to detect boundares we have to blurr it first 
#blurr works on a gray scale : store intensity of each pixels 
orig=img.copy()
gray=cv2.cvtColor(orig,cv2.COLOR_BGR2GRAY)
blurred=cv2.GaussianBlur(gray,(5,5),0)
# here 0 is  a sigma standard deviation 
plt.imshow(gray)
plt.show()
print("---------this is gray scale image-------- ")
plt.imshow(blurred)
plt.show()
print("---------this is blurred  scale image-------- ")
print(gray.shape)
print("---gray shape----")
print(blurred.shape)
print("---blurred shape----")
```
